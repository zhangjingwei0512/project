```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import datetime
from pandas_datareader import data
import requests
from bs4 import BeautifulSoup
from matplotlib import pyplot as plt
import seaborn as sns
import csv
```

```python
start=datetime.datetime(1989,1,1)
end=datetime.date.today()
```

```python
sp=web.DataReader("^GSPC","yahoo",start,end)
```

```python
sp
```

```python
%matplotlib inline
%pylab inline
```

```python
HV=(sp['Close']-sp['Close'])/sp['Close']
for j in range(0,7540):
    HV[j]=(sp['Close'][j]-sp['Close'][j-1])/sp['Close'][j-1]
HV1=HV[1:]
HV1
```

```python
HV1.sort_values(ascending=True).head(10)
```

```python
sp["Close"].plot(grid = True)
```

```python
df=pd.read_csv('MEILIANCHU.csv')
df
```

```python
from datetime import datetime
from dateutil import parser
import numpy
def parse_datetime(x):
    try:
        return parser.parse(x)
    except:
        return numpy.nan
df['datetime'] = df['Time Period'].apply(parse_datetime)
df 
```

```python
df.set_index('datetime').resample('m').aggregate('max').plot()
```
